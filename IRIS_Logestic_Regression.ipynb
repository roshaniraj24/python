```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import OrdinalEncoder,OneHotEncoder,StandardScaler,MinMaxScaler,LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression
```

```python
df=pd.read_csv('iris.csv')
df
```

```python
df.isnull().sum()
```

```python
x=df.iloc[:,0:3]
x
```

```python
y=df.iloc[:,-1]
y
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.8,random_state=0)
```

```python
xtrain
```

```python
xtest
```

```python
ytrain
```

```python
ytest
```

```python
le=LabelEncoder()
y_tr=le.fit_transform(ytrain)
```

```python
y_te=le.fit_transform(ytest)
```

```python
log=LogisticRegression()
```

```python
log.fit(xtrain,y_tr)
```

```python
y_pred=log.predict(xtest)
y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
print("Accuracy score =",accuracy_score(y_te,y_pred))
```

```python
pd.DataFrame(confusion_matrix(y_te,y_pred))
```

```python
result=pd.DataFrame()
result['Actual Label']=ytest
result['Decision Tree prediction']=y_pred
```

```python
result
```

```python
from sklearn.metrics import precision_score
precision_score(y_te,y_pred,average='weighted')
```

```python
from sklearn.metrics import recall_score
recall_score(y_te,y_pred,average='micro')
```

```python
from sklearn.metrics import f1_score
f1_score(y_te,y_pred,average='micro')
```
