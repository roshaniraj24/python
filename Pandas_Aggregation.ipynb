# PANDAS_PIVOT_AND_MELT_FUNCTION

```python
import pandas as pd
from pandas import DataFrame
df1=DataFrame(
    [
             ['Aarohi','Singh','Full-time','adminstration',2,20000],
             ['Arpita','Agarwal','intern','Technical',3,5000],
             ['Riya','shah','Full-time','administration',5,100000],
             ['Yash','Bhatia','Part-time','Management',6,20000],
             ['Priya','Alian','full-time','Technical',6,20000],
    ], columns=['First name','Last name','Type','Department','YoE','Salary']
             )
df1
```

```python
table = pd.pivot_table(df1,index="Last name")
table
```

```python
table = pd.pivot_table(df1,index="Type")
table
```

```python
#Find mean value of salary
table = pd.pivot_table(df1,index="Type",values="Salary")

table
```

```python
p_table = pd.pivot_table(df1,index="Type", columns="Department",values="Salary")
p_table
```
