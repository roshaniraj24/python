```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import OrdinalEncoder,OneHotEncoder,StandardScaler,MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression
```

```python
df=pd.read_csv('titanic.csv')
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.drop(['PassengerId','Ticket','Fare','Cabin','Name'], inplace=True , axis=1)
```

```python
df.fillna(method='ffill' , inplace=True)
```

```python
df.isnull().sum()
```

```python
x=df.iloc[:,1:]
```

```python
x
```

```python
y=df.iloc[:,0]
y
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2 , random_state=2)
```

```python
xtrain
```

```python
xtest
```

```python
ytrain
```

```python
ytest
```

```python
tr = ColumnTransformer(transformers=[
    ('t2',OneHotEncoder(drop='first',sparse=False),['Sex','Embarked'])
] , remainder='passthrough')
```

```python
a = tr.fit_transform(xtrain)
a
```

```python
b = tr.fit_transform(xtest)
b
```

```python
log = LogisticRegression()
```

```python
log.fit(a,ytrain)
```

```python
y_pred = log.predict(b)
y_pred
```

```python
from sklearn.metrics import accuracy_score , confusion_matrix
```

```python
print("Accuracy matrix =",accuracy_score(ytest,y_pred))
```

```python
pd.DataFrame(confusion_matrix(ytest , y_pred))
```

```python
result=pd.DataFrame()
result['Actual Label'] = ytest
result['Decision Tree prediction'] = y_pred
```
