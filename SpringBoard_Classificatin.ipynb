# For the Bank Marketing dataset available in UCI Machine Learning Repository, induct a decision tree to predict whether a client will subscribe to a term deposit or not (given in column 'y'). 

# 1.. Consider the column 'y' as the target variable. From the remaining columns, consider all except 'duration' as predictors

```python
import numpy as np
import pandas as pd
```

```python
a=pd.read_csv("bankadditionalfull1603958996760.zip")
a
```

```python
a['job'].unique()
```

```python
a.describe()
```

```python
a.isnull().sum()
```

# 2..Encode the predictors to convert them to numerical data.

```python
x=a.drop(columns=['duration','y'])
x
```

```python
y=a.iloc[:,-1:]
y
```

```python
x=pd.get_dummies(x,drop_first=True)
x
```

# 3..Split the data into train and test with a ratio of 80:20.

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=20,random_state=0)
```

```python
xtrain
```

```python
xtest
```

```python
ytrain
```

```python
ytest
```

# 4..Split the data into train and test with a ratio of 80:20.

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
new=DecisionTreeClassifier()
```

```python
new.fit(xtrain,ytrain)
```

```python
pred=new.predict(xtest)
pred
```

```python
ytest
```

# 5..Determine the accuracy, precision and recall for every class in the outcome.

```python
from sklearn.metrics import precision_score,recall_score,accuracy_score,confusion_matrix
```

```python
accuracy_score(ytest,pred)
```

```python
precision_score(ytest,pred,average="micro")
```

```python
confusion_matrix(ytest,pred)
```

```python
recall_score(ytest,pred,average="micro")
```
